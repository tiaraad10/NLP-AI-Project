# Question Answering System
---
---
Penggunaan metode open domain extractive Question Answering System dilakukan setelah melihat limitasi yang diberikan bila menggunakan metode pengambilan jawaban yang mirip menggunakan persamaan semantik. Dalam metode ini, QAS hanya mengambil jawaban yang optimal dari suatu konteks yang diberikan. Konteks yang diberikan ini merupakan kumpula jawaban yang dapat menceritakan atau mendeskripsikan suatu konten ataupu objek yang akan ditanyakan.

Adapun macam Question Answering System berdasarkan penggunaan data, yaitu:


1.   Open Domain
> Membutuhkan konteks yang telah ada sebagai bahan acuan pemberi jawaban. (Pelatihan model hanya mempelajari pola pemberian jawaban)



2.   Closed Domain
> Tidak membutuhkan konteks pada saat pengguna bertanya (Model telah terlatih dengan konteks yang diberikan pada saat pelatihan model)

Dalam aspek metode yang digunakan, maka Question Answering System terbagi menjadi:


1.   Extractive QA
> Menggunakan metode retrieval-reader untuk melakukan pembacaan konteks data yang diberikan lalu diambil jawaban yang paling relevan


2.   Abstractive QA
> Lebih menggunakan metode text-generation dimana model akan membuat jawaban untuk pertanyaan yang dilontarkan oleh pengguna. Tentunya model ini cendrung memiliki komputasi yang sangat besar. Akan tetapi jawaban yang diberikan bukan hanya sekedar mengambil jawaban dari konteks, tetapi menggunakan kalimat yang memimik cara penulisan manusia sehingga terkesan lebih natural

```python
from transformers import BertTokenizer, BertForQuestionAnswering
import torch
import re
import pandas as pd
import numpy as np

model_name = '/content/drive/MyDrive/PA_NLP/best_model'

tokenizer = BertTokenizer.from_pretrained(model_name)
model = BertForQuestionAnswering.from_pretrained(model_name)
```

```python
def normalize_text(text):
  text = text.lower().strip()
  text = re.sub(r'https?://\S+|www\.\S+', '', text)
  text = re.sub(r'[-+]?[0-9]+', '', text)
  text = re.sub(r'[^\w\s.,]','', text)
  return text


def ask(question, context):
  # Init
  answer = ""

  # Normalisasi
  question_normalized = normalize_text(question)

  # Tokenisasi - Convert string
  input_ids = tokenizer.encode(
      question_normalized,
      context,
  )

  tokens_ids_converted = tokenizer.convert_ids_to_tokens(input_ids)

  # Ambil seperator untuk memisahkan question dan context
  separator_ids = input_ids.index(tokenizer.sep_token_id)

  segment_text_q = separator_ids + 1
  segment_text_a = len(input_ids) - segment_text_q

  segmented_text_ids = [0]*segment_text_q + [1]*segment_text_a

  # Pengecekan segmentasi question dan context menghasilkan panjang tokenisasi yang sama
  assert len(segmented_text_ids) == len(input_ids)

  # Prediksi
  outputs = model(
      torch.tensor([input_ids]),
      token_type_ids = torch.tensor([segmented_text_ids])
  )

  answer_start = torch.argmax(outputs.start_logits)
  answer_end = torch.argmax(outputs.end_logits)

  if answer_end >= answer_start:
    answer = tokens_ids_converted[answer_start]

    for i in range(answer_start+1, answer_end+1):
      if tokens_ids_converted[i][0:2] == "##":
        answer += tokens_ids_converted[i][2:]
      else:
        answer += " " + tokens_ids_converted[i]


  if answer.startswith("[CLS]"):
    answer = "Jawaban tidak ditemukan"

  return answer.capitalize()
```

```python
from transformers import BertConfig, DataCollatorWithPadding
from transformers.models.bert.modeling_bert import BertEncoder


def weights_init(teacher_model, student_model):
  if isinstance(teacher_model, BertModel) or type(teacher_model).__name__.startswith('BertFor'):
    for t, s in zip(teacher_model.children(), student_model.children()):
      weights_init(t, s)

  elif isinstance(teacher_model, BertEncoder):
    teacher_encode_l = [i for i in next(teacher_model.children())]
    student_encode_l = [i for i in next(student_model.children())]

    for i in range(len(student_encode_l)):
      student_encode_l[i].load_state_dict(teacher_encode_l[6*i].state_dict())

  else:
    student_model.load_state_dict(teacher_model.state_dict())


def model_optimization(teacher_model):
  configuration_dict = teacher_model.config.to_dict()
  configuration_dict['num_hidden_layers'] //= 6
  configuration_dict = BertConfig.from_dict(configuration_dict)


  student_model = type(teacher_model)(configuration_dict)
  weights_init(teacher_model, student_model)
  return student_model
```

```python
distil_model = model_optimization(model)
```

## Dataset

```python
def preprocess_data(qas_data):
  texts = []
  queries = []
  answers = []

  for group in qas_data['data']:
    for passage in group['paragraphs']:
      context = passage['context']
      for qa in passage['qas']:
        question = qa['question']
        for answer in qa['answers']:
          texts.append(context)
          queries.append(question)
          answers.append(answer)

  return texts, queries, answers


def preprocess_data_given_star_end(answers, texts):
  for answer, text in zip(answers, texts):
    real_answer = answer['text']
    start_idx = answer['answer_start']

    end_idx = start_idx + len(real_answer)

    if text[start_idx:end_idx] == real_answer:
        answer['answer_end'] = end_idx

    elif text[start_idx-1:end_idx-1] == real_answer:
        answer['answer_start'] = start_idx - 1
        answer['answer_end'] = end_idx - 1

    elif text[start_idx-2:end_idx-2] == real_answer:
        answer['answer_start'] = start_idx - 2
        answer['answer_end'] = end_idx - 2

def add_token_positions(encodings, answers):
  start_positions = []
  end_positions = []

  count = 0

  for i in range(len(answers)):
    start_positions.append(encodings.char_to_token(i, answers[i]['answer_start']))
    end_positions.append(encodings.char_to_token(i, answers[i]['answer_end'] - 1))


    if start_positions[-1] is None:
      start_positions[-1] = tokenizer.model_max_length


    if end_positions[-1] is None:
      end_positions[-1] = tokenizer.model_max_length

      # if end_positions[-1] is None:
      count += 1
      #   end_positions[-1] = tokenizer.model_max_length

  print(count)

  # Update the data in dictionary
  encodings.update({'start_positions': start_positions, 'end_positions': end_positions})


class SquadDataset(torch.utils.data.Dataset):
    def __init__(self, encodings):
        self.encodings = encodings

    def __getitem__(self, idx):
        return {key: torch.tensor(val[idx]) for key, val in self.encodings.items()}

    def __len__(self):
        return len(self.encodings.input_ids)
```

```python
texts_train, queries_train, answers_train = preprocess_data(data_train)
texts_test, queries_test, answers_test = preprocess_data(data_test)

preprocess_data_given_star_end(answers_train, texts_train)
preprocess_data_given_star_end(answers_test, texts_test)


train_encodings = tokenizer(
    texts_train,
    queries_train,
    truncation=True,
    padding=True,
    max_length=128,
)
val_encodings = tokenizer(
    texts_test,
    queries_test,
    truncation=True,
    padding=True,
    max_length=128,
)


add_token_positions(train_encodings, answers_train)
add_token_positions(val_encodings, answers_test)

train_dataset = SquadDataset(train_encodings)
val_dataset = SquadDataset(val_encodings)
```

## Training

```python
from transformers import TrainingArguments, Trainer
```

```python
import os

dir_saved = '/content/drive/MyDrive/PA_NLP/Torche Intern NLP Progress/DistilledQASBert'
if not os.path.exists(dir_saved):
  os.mkdir(dir_saved)
```

```python
class CustomTrainingArguments(TrainingArguments):
    def __init__(self, *args, alpha=0.5, temperature=2.0, **kwargs):
        super().__init__(*args, **kwargs)

        self.alpha = alpha
        self.temperature = temperature

class CustomTrainer(Trainer):
    def __init__(self, *args, teacher_model=None, **kwargs):
        super().__init__(*args, **kwargs)
        self.teacher = teacher_model

        self._move_model_to_device(self.teacher,self.model.device)
        self.teacher.eval()

    def compute_loss(self, model, inputs, return_outputs=False):
        outputs_student = model(**inputs)
        student_loss=outputs_student.loss

        with torch.no_grad():
          outputs_teacher = self.teacher(**inputs)


        assert outputs_student.logits.size() == outputs_teacher.logits.size()


        loss_function = torch.nn.MSELoss(reduction='sum')
        loss = loss_function(student_loss,outputs_student) * (self.args.temperature ** 2)

        loss = self.args.alpha * loss + (1. - self.args.alpha)
        return (loss, outputs_student) if return_outputs else loss
```

```python
os.environ["WANDB_DISABLED"] = "true"

device = torch.device('cuda') if torch.cuda.is_available() else torch.device('cpu')
model.to(device)
model.train()

args = CustomTrainingArguments(
    output_dir = '/content/output_dir',
    overwrite_output_dir = True,
    per_device_train_batch_size = batch_size,
    per_device_eval_batch_size=int(batch_size/4),
    num_train_epochs = 200,
    optim = 'adamw_torch',
    learning_rate=1e-5,
    weight_decay=0.01,
)


trainer = CustomTrainer(
    distil_model,
    args,
    teacher_model=model,
    train_dataset=train_dataset,
    eval_dataset=val_dataset,
)



trainer.train()

trainer.save_model(dir_saved)
```
